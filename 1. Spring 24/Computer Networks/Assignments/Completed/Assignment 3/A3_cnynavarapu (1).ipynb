```python
from scapy.layers.dns import DNS, DNSQR
from scapy.layers.inet import UDP, TCP
from scapy.layers.inet6 import IPv6, ICMPv6EchoRequest, ICMPv6EchoReply, ICMPv6TimeExceeded
from scapy.layers.l2 import Ether
from scapy.all import *

# creating new packets
new_packets = []

# Creating a new packet stacked with Ether, IPv6, TCP, and a payload
packet_1 = (Ether(src='34:16:12:1c:3c:69', dst='94:ce:55:8c:fa:c5') /
            IPv6(src='2001:bdbd:fa36:938f:3436:2e9d:c8cd:28eb', dst='2001:26f2:3eb0:1eb6:7152:c40f:4696:ca50') /
            UDP(sport=38763, dport=18327) /
            b"67165ea06bd8d2726290")
new_packets.append(packet_1)

# creating a new packet stacked with Ethernet, IPv6, UDP, and a payload
packet_2 = (Ether(src='98:3d:eb:7f:12:ea', dst='34:16:12:1c:3c:69') /            
            IPv6(src='2001:bdbd:fa36:938f:f570:ac62:41df:eca8', dst='2001:bdbd:fa36:938f:3436:2e9d:c8cd:28eb') /
            ICMPv6EchoRequest())
new_packets.append(packet_2)

# creating a new packet stacked with Ether, IPv6, ICMPv6
packet_3 = (Ether(src='94:ce:55:8c:fa:c5', dst='98:3d:eb:7f:12:ea') /
            IPv6(src='2001:26f2:3eb0:1eb6:7152:c40f:4696:ca50', dst='2001:bdbd:fa36:938f:f570:ac62:41df:eca8') /
            UDP(sport=35931, dport=25561) /
            b"aee12d346eb954541")
new_packets.append(packet_3)

# creating a new packet stacked with Ethernet, IPv6, UDP, and a payload
packet_4 = (Ether(src='94:ce:55:8c:fa:c5', dst='b0:c2:0a:3d:af:4e') /
            IPv6(src='2001:26f2:3eb0:1eb6:7152:c40f:4696:ca50', dst='2001:bdbd:fa36:938f:adfb:39a4:60e7:b299') /
            UDP(sport=51046, dport=12978) /
            b"c41279c5e19e")
new_packets.append(packet_4)

# creating a new packet stacked with Ethernet, IPv6, UDP, and a payload
packet_5 = (Ether(src='98:3d:eb:7f:12:ea', dst='94:ce:55:8c:fa:c5') /
            IPv6(src='2001:bdbd:fa36:938f:f570:ac62:41df:eca8', dst='2001:26f2:3eb0:1eb6:7152:c40f:4696:ca50') /
            UDP(sport=9616, dport=32968) /
            b"3ed68a43d92c881cefaaa")
new_packets.append(packet_5)

# creating a new packet stacked with Ether, IPv6, ICMPv6************
packet_6 = (Ether(src='94:ce:55:8c:fa:c5', dst='34:16:12:1c:3c:69') /
            IPv6(src='2001:26f2:3eb0:1eb6:7152:c40f:4696:ca50', dst='2001:bdbd:fa36:938f:3436:2e9d:c8cd:28eb') /
            TCP(sport=7992, dport=29656) /
            b"a392967c4d838a04ef035a")
            
new_packets.append(packet_6)

# creating a new packet stacked with Ethernet, IPv6, DNS, and a payload
payload7_hex = b'b2732f8f001e000063626534643066393331366362383535333931343738'
payload_bytes = bytes.fromhex((payload7_hex.strip().replace(b'\n', b'').decode()))
packet_7 = (Ether(src='34:16:12:1c:3c:69', dst='94:ce:55:8c:fa:c5') /
             IPv6(src='2001:bdbd:fa36:938f:3436:2e9d:c8cd:28eb', dst='2001:26f2:3eb0:1eb6:7152:c40f:4696:ca50') /
             ICMPv6TimeExceeded(code=0) /
             Raw(load=payload_bytes))
new_packets.append(packet_7)

# creating new packet stacked with Ethernet, IPv6, UDP, and a payload
packet_8 = (Ether(src='34:16:12:1c:3c:69', dst='94:ce:55:8c:fa:c5') /
            IPv6(src='2001:bdbd:fa36:938f:3436:2e9d:c8cd:28eb', dst='2001:26f2:3eb0:1eb6:7152:c40f:4696:ca50') /
            UDP(sport=56756, dport=1782) /
            b"29c39136bac42c646dc2cb2227e")
new_packets.append(packet_8)

# Creating a new packet stacked with Ether, IPv6, TCP, and a payload
packet_9 = (Ether(src='34:16:12:1c:3c:69', dst='94:ce:55:8c:fa:c5') /
            IPv6(src='2001:bdbd:fa36:938f:3436:2e9d:c8cd:28eb', dst='2001:26f2:3eb0:1eb6:7152:c40f:4696:ca50') /
            TCP(sport=63632, dport=53) /
            DNS(rd=1, qd=DNSQR(qname="www.example.com", qtype="A"))
            )
new_packets.append(packet_9)

# creating a new packet stacked with Ethernet, IPv6, ICMPv6
packet_10 = (Ether(src='34:16:12:1c:3c:69', dst='98:3d:eb:7f:12:ea') /
            IPv6(src='2001:bdbd:fa36:938f:3436:2e9d:c8cd:28eb', dst='2001:bdbd:fa36:938f:f570:ac62:41df:eca8') /
            ICMPv6EchoReply())
new_packets.append(packet_10)

# Creating a new packet stacked with Ether, IPv6, TCP, and a payload
packet_11 = (Ether(src='94:ce:55:8c:fa:c5', dst='98:3d:eb:7f:12:ea') /
            IPv6(src='2001:26f2:3eb0:1eb6:7152:c40f:4696:ca50', dst='2001:bdbd:fa36:938f:f570:ac62:41df:eca8') /
            UDP(sport=31055, dport=5099) /
            b"81f731be5b01b")
new_packets.append(packet_11)

# creating a new packet stacked with Ethernet, IPv6, UDP, and a payload
packet_12 = (Ether(src='94:ce:55:8c:fa:c5', dst='34:16:12:1c:3c:69') /
             IPv6(src='2001:26f2:3eb0:1eb6:7152:c40f:4696:ca50', dst='2001:bdbd:fa36:938f:3436:2e9d:c8cd:28eb') /
             UDP(sport=32735, dport=56613) /
             b"c31f48da8d70ed6c330410e1f46")
new_packets.append(packet_12)

# Populating timestamps
timestamps = [0, 0.1, 0.064235000, 0.084097000, 0.110350000, 0.140054000, 0.163191000, 0.180356000, 0.222465000 0.235771000, 0.284292000, 0.291182000]

# Assigning timestamps to packets
for ts, pkt in zip(timestamps, new_packets):
    pkt.time = ts

# Saving packets to a pcap file
wrpcap('A3_cnynavarapu.pcap', new_packets)

# Reading pcap files
pcap1 = rdpcap("cnynavarapu.pcap")
pcap2 = rdpcap("A3_cnynavarapu.pcap")

# Comparing packets
for packet1 in pcap1:
    if packet1 not in pcap2:
        print(f"Packet {packet1.summary()} from file1.pcap was dropped")
```
